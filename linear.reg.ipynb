```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
print(iris.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='species', y='sepal_length', data=iris, hue='species', palette="viridis", dodge=False)
plt.title("Average Sepal Length per Species", fontsize=14)
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.legend([],[], frameon=False)  # Hide legend
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(iris['sepal_length'], bins=30, kde=True, color="blue")
plt.title("Distribution of Sepal Length", fontsize=14)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(iris, x="sepal_length", hue="species", bins=20, kde=True, palette="Set1")
plt.title("Distribution of Sepal Length by Species", fontsize=14)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.show()
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(12, 8))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(iris[feature], bins=20, kde=True, color="green")
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='sepal_length', data=iris, palette="coolwarm")
plt.title("Box Plot of Sepal Length by Species", fontsize=14)
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
sns.pairplot(iris, hue="species", palette="husl")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='species', y='sepal_length', data=iris, hue='species', palette="muted", dodge=False)
plt.title("Violin Plot of Sepal Length by Species", fontsize=14)
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.legend([],[], frameon=False)  # Hide legend
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap", fontsize=14)
plt.show()
```

```python
#model2
# dependent variable
y = iris[['sepal_length']] 
# multiple linear regersiion
x = iris[['sepal_width','petal_length', 'petal_width']]
```

```python
x_train, x_test , y_train, y_test = train_test_split(x,y,test_size = 0.3)
```

```python
lr2 = LinearRegression()
```

```python
lr2.fit(x_train, y_train)
```

```python
y_pred =lr2.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```

```python
plt.scatter(y_test, y_test - y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Residuals (Errors)")
plt.title("Residual Plot")
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted")
plt.show()
```
